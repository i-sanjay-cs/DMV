```python
import pandas as pd
```

```python
df = pd.read_csv('city_day.csv')
```

```python
df
```

```python
df.dtypes
```

```python
Date = df['Date']
AQI_values = df['AQI']
Pollutant_level = df[['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(Date,AQI_values,c='r')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(Date,Pollutant_level)
plt.legend()
plt.show()
```

```python
plt.bar(df['Date'], df['AQI'], label="AQI")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.title("AQI Comparison Across Dates")
plt.legend()
plt.show()
```

```python
plt.box()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df["PM2.5"], df['AQI'], label="PM2.5", alpha=0.5)
plt.scatter(df["PM10"], df[AQI_values], label="PM10", alpha=0.5)
plt.scatter(df["CO"], df[AQI_values], label="CO", alpha=0.5)
plt.xlabel("Pollutant Levels")
plt.ylabel("AQI")
plt.title("Scatter Plot of AQI vs Pollutant Levels")
plt.legend()
plt.show()
```
